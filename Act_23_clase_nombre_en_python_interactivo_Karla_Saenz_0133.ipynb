```python

class Entidad:
    def __init__(self, id_code, nombre, valor, categoria):
        self.id_code = id_code
        self.nombre = nombre
        self.valor = valor
        self.categoria = categoria

    def show_vibe(self):
        print("\n✨ --- BACKSTAGE: DATOS RECUPERADOS --- ✨")
        print(f"🆔 ID/Código: {self.id_code}")
        print(f"👤 Nombre: {self.nombre}")
        print(f"💰 Valor: {self.valor}")
        print(f"🏷️ Categoría: {self.categoria}")
        print("------------------------------------------")

# --- La Entrevista (Interacción) ---
print("🎤 ¡Hey! Necesito que me des los datos para el show:")

id_input = input("1. Dame el Código (ID): ")
nom_input = input("2. ¿Cómo se llama la estrella? (Nombre): ")
val_input = input("3. ¿Cuál es su Valor/Precio?: ")
cat_input = input("4. ¿En qué Categoría brilla?: ")

# Creando el objeto estrella
mi_objeto = Entidad(id_input, nom_input, val_input, cat_input)

# ¡Que ruede el show!
mi_objeto.show_vibe()
```
